```python
#below is code for some of the error handling

class Invalid_Game_Choice(Exception):
    pass
class Invalid_RPS_Choice(Exception):
    pass
class Invalid_Diff_Choice(Exception):
    pass

#below is code for the Virtual Arcade

def Play_Virtual_Arcade():
    #the code below welcomes the user into the virtual arcade and provides an option to leave
    welcome = input("Welcome to Virtual Arcade! Enter anything to continue or quit to leave.")
    if welcome == 'quit':
        print("Goodbye.")
        return
    else:
        name = input("Enter your name")
        #Here, the user is asked for their name and subsequently the arcade difficulty they wish to choose while also handling invalid input
        while True:
            try:
                diff = input("Choose a difficulty from easy, medium, or hard. Easy starts with 3 lives, medium with 2, and hard with 1.")
                if diff not in ["easy", "medium", "hard"]:
                    raise Invalid_Diff_Choice
                    continue
                else:
                    break
            except Invalid_Diff_Choice:
                print("Invalid difficulty choice. Try again.")  
        diff = diff.lower()
        win_total = 0
        #Here, the user is assigned an amount of lives corresponding to their selected difficulty
        if diff == 'easy':
            print("You have chosen easy and will start with 3 lives")
            lives = 3
        elif diff == 'medium':
            print("You have chosen medium and will start with 2 lives")
            lives = 2
        elif diff == 'hard':
            print("You have chosen hard and will start with 1 lives")
            lives = 1
        while True:
            try:
                #The code below runs the framework of the virtual arcade. Also, the user is asked to choose which mini game they wish to play
                if lives == 0:
                    print(f"Game over, you are out of lives. You scored {win_total} at the Virtual Arcade. Thank you for playing {name}!")
                    break
                choose_game = input("Enter the number for one of the following mini games. 1 for a word game, 2 for Trivia, 3 for rock, paper, scissors")
                #The code below executes the calling of each minigame to run depending on the users selection
                #The score and lives are updated accordingly based on user performance
                if choose_game == '1':
                    if play_word_game() == True:
                        win_total = win_total + 1
                        continue
                    else:
                        lives = lives - 1
                        continue
                if choose_game == '2':
                    if play_trivia() == True:
                        win_total = win_total + 1
                        continue
                    else:
                        lives = lives - 1
                        continue
                if choose_game == '3':
                    if RPS() == True:
                        win_total = win_total + 1
                        continue
                    else:
                        lives = lives - 1
                        continue
                else:
                    #Here, the error of invalid user input for the mini game choice is handled
                    raise Invalid_Game_Choice
                    continue
            except Invalid_Game_Choice:
                print("Invalid game choice, try again.")             
def RPS():
    #The code below sets up the required components for rock, paper, scissors such as importing the randint function from random
    #The user is also welcomed into the minigame
    from random import randint
    win_count = 0
    loss_count = 0
    message = input("Welcome to Rock, Paper, Scissors!. You must win 4 games before the bot does in order to win. Enter anything to continue")
    while True:
        
        #check status of series
        if win_count == 4:
            print(f"Congratulations, you have won the series {win_count} to {loss_count}")
            return True
            break
        if loss_count == 4:
            print(f"Unfortunatley, you have lost the series {win_count} to {loss_count}")
            return False
            break
        while True:
            try:
                #The code below asks the user to choose rock, paper, or scissors and handles invalid user input
                user = input("Enter rock, paper, or scissors. Enter quit to end the game and lose a life")
                if user not in ['rock', 'paper', 'scissors', 'quit']:
                    raise Invalid_RPS_Choice
                    continue
                else:
                    break
            except Invalid_RPS_Choice:
                print("Invalid choice, you must enter rock, paper, or scissors. Try again.")
        #Here, the user input is made lowercase to be standardized for efficent usage and the computer choice is randomly generated
        user = user.lower()
        Bot = randint(0,2)
        
        #if user entered quit, this code allows them to leave but it will cost a life
        if user == "quit":
            print("Goodbye, you will return to the main arcade")
            return False
            break
            
        #test against bot as rock
        if Bot == 0:
            print("Bot chooses rock")
            if user == 'rock':
                print(f"draw, you have{win_count} and the bot has {loss_count} in the series")
                continue
            elif user == 'paper':
                win_count = win_count + 1
                print(f"Nice win, you have{win_count} and the bot has {loss_count} in the series")
                continue
            elif user == "scissors":
                loss_count = loss_count + 1 
                print(f"Bot wins, you have{win_count} and the bot has {loss_count} in the series" )
                continue
                
        #test against bot as paper
        elif Bot == 1:
            print("Bot chooses paper")
            if user == 'rock':
                loss_count = loss_count + 1
                print(f"Bot wins, you have{win_count} and the bot has {loss_count} in the series")
                continue
            elif user == 'paper':
                print(f"draw, you have{win_count} and the bot has {loss_count} in the series")
                continue
            elif user == 'scissors':
                win_count = win_count + 1
                print(f"Nice win, you have{win_count} and the bot has {loss_count} in the series")
                continue
            
       #test against bot as scissors
        elif Bot == 2:
            print("Bot chooses scissors")
            if user == 'rock':
                win_count = win_count + 1
                print(f"Nice win, you have{win_count} and the bot has {loss_count} in the series")
                continue
            elif user == 'paper':
                loss_count = loss_count + 1
                print(f"Bot wins, you have{win_count} and the bot has {loss_count} in the series")
                continue
            elif user == 'scissors':
                print(f"draw, you have{win_count} and the bot has {loss_count} in the series")
                continue
    return

#Below is the code for the word game

def is_valid_guess(guess):
    # Checks if the guess is a single letter
    result = len(guess) == 1 and guess.isalpha()
    return result

def play_word_game():#the user chooses a difficulty and invalid user input is handled
    while True:
        try:
            word_game_level = input("Pick a difficulty you would like to play at. Your choices are easiest, easy, medium, hard, and hardest")
            if word_game_level == "easiest" or word_game_level == "easy" or word_game_level == "medium" or word_game_level == "hard" or word_game_level == "hardest":
                break
            else:
                raise ValueError("Invalid choice, pick again")

        except ValueError as e:
            print(e)

    #below are the words given based on the choice of difficulty the user chooses. Automatically chooses easiest if incorrect input
    if word_game_level == "easiest":
        word_to_guess = "sleep"
    elif word_game_level == "easy":
        word_to_guess = "foolishness"
    elif word_game_level == "medium":
        word_to_guess = "jumping rope"
    elif word_game_level == "hard":
        word_to_guess = "the man fell asleep while jumping rope"
    elif word_game_level == "hardest":
        word_to_guess = "the zebra named jack is quiet"
    else: 
        word_to_guess = "sleep"

    print ("You have 6 guesses to complete this word")
    #these variables will be used in the while loop to track different aspects of the game
    #the variable below makes a string that is the same length as the word to guess of all ?s exept where there are spaces
    current_word = ""
    for x in word_to_guess:
        if x == ' ':
            current_word += ' ' 
        else:
            current_word += '?'  
    
    incorrect_guesses = []
    number_of_guesses_left = 6
    while '?' in current_word and number_of_guesses_left > 0: #the game continues until you win or run out of guesses
        print(current_word)
        try: #below are errors that ensure each guess is a single letter and that there are no repeats
            player_guess = input("What letter do you guess? ").lower()
            if not is_valid_guess(player_guess):
                raise ValueError("Invalid guess. Please enter a single letter.")
            
            if player_guess in incorrect_guesses or player_guess in current_word:
                raise ValueError("You already guessed that letter. Try again!")
            
            if player_guess in word_to_guess:
                updated_word = ""

            #below is a for loop that replaces the current_hangmans question marks with correct guesses
                for i in range(len(word_to_guess)):
                    if word_to_guess[i] == player_guess:
                        updated_word += player_guess
                    else:
                        updated_word += current_word[i]
                current_word = updated_word
            else:
                print("Incorrect guess. Try again!")
                incorrect_guesses.append(player_guess) #tracks the incorrect guess letters
            print("Incorrect guesses:", incorrect_guesses)
            number_of_guesses_left = 6 - len(incorrect_guesses) #tracks the number of guesses left
            print("Incorrect guesses left before game ends: ", number_of_guesses_left)
        except ValueError as e:
            print(e)
    
        if current_word == word_to_guess: #winning condition
            print ("Congratulations you won!")
            return True
        elif number_of_guesses_left == 0: #losing condition
            print ("You have run out of attempts, feel free to replay the game")
            return False
        
#Below is the code for trivia

def is_valid_answer(answer):
    # Checks if the guess is a string
    result = answer.isalpha()
    
    return result

def play_trivia():#the user chooses a difficulty and invalid user input is handled
    while True:
        try:
            trivia_level = input("What level do you want for Geography trivia? Your choices are easiest, easy, medium, hard, and hardest")
            print("You are allowed 3 incorrect guesses")
            if trivia_level == "easiest" or trivia_level == "easy" or trivia_level == "medium" or trivia_level == "hard" or trivia_level == "hardest":
                break
            else:
                raise ValueError("Invalid choice, pick again")

        except ValueError as e:
            print(e)

    #below are the words given based on the choice of difficulty the user chooses
    if trivia_level == "easiest":
        print ("What country is Paris located in?")
        word_to_guess = "france"
    elif trivia_level == "easy":
        print ("What country is Cancun located in?")
        word_to_guess = "mexico"
    elif trivia_level == "medium":
        print ("What country is Brussels located in?")
        word_to_guess = "belgium"
    elif trivia_level == "hard":
        print ("What is the biggest country that has the Himalaya mountain range?")
        word_to_guess = "china"
    elif trivia_level == "hardest":
        print ("What country is Lake Paranoa located in?")
        word_to_guess = "brazil"
    else: 
        print ("What country is Paris located in?")
        word_to_guess = "france"
        
    amount_of_guesses=0
    user_guess = "blank"
    while amount_of_guesses < 3 and word_to_guess != user_guess:
        try: #below handles errors to ensure each guess is a string
            user_guess = input("What is your answer").lower()
            if not is_valid_answer(user_guess):
                raise ValueError("Invalid answer. Please enter a word with only letters")
            amount_of_guesses += 1
            if user_guess != word_to_guess:
                print ("Incorrect, try again")
        except ValueError as e:
            print(e)
        if user_guess==word_to_guess:
            #Here, the winning condition is satisfied, the user is notified of this and True is returned
            print ("Congratulations, you got the correct answer!")
            return True
        elif amount_of_guesses > 2:
            #Here, the user used all incorrect attempts, the user is notified of this and False is returned
            print ("You did not get the correct answer")
            return False
    
Play_Virtual_Arcade()
```
